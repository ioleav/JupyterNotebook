```python
municipios_extremadura_shp["area"] = municipios_extremadura_shp.area
municipios_extremadura_shp.head(3)
```

```python
municipios_extremadura_shp["area"]
```

```python
municipios_extremadura_shp["perimetro"] = municipios_extremadura_shp.boundary
municipios_extremadura_shp.head(3)
#calcular el perimetro
```

```python
municipios_extremadura_shp ["centroide"] = municipios_extremadura_shp.centroid
municipios_extremadura_shp.head(3)
#calcular el centroide
```

```python
#calcular la distancia de un punto A a un punto B
punto_inicial = municipios_extremadura_shp ["centroide"].iloc[0] 
#busca los registros por un indice aqui le estamos diciendo que empiece desde 0 .iloc[0], por lo cual, 
#se refiere al primer registro
```

```python
punto_inicial
```

```python
municipios_extremadura_shp["distancia"] = municipios_extremadura_shp ["centroide"].distance(punto_inicial)
municipios_extremadura_shp["distancia"]
```

```python
municipios_extremadura_shp["distancia"].mean()
```

```python
#SETEAR LA GEOMETRIA A POLIGONOS
municipios_extremadura_shp = municipios_extremadura_shp.set_geometry("geometry")
municipios_extremadura_shp.plot()
```

```python
municipios_extremadura_shp.plot(column = "NOMBRE03", legend = True, legend_kwds = {"loc": "upper left"},figsize = (20,50))
```

```python
municipios_extremadura_shp["envolvente"] = municipios_extremadura_shp.convex_hull
#metodo convex hull es para suavizar los poligonos, se creara una geometria nueva a la que hemos llamado envolvente
#nos genera esa columna con una geometria que envuelve de forma mas simplificada cada uno de los polgionos
```

```python
municipios_extremadura_shp["envolvente"].plot(alpha = 0.5)
#alpha es para la transferencia
```

```python
axis = municipios_extremadura_shp["envolvente"].plot(alpha = 0.5)
municipios_extremadura_shp["perimetro"].plot (ax = axis, color = "white", linewidth = 0.5)
```

```python
municipios_extremadura_shp["buffer"] = municipios_extremadura_shp.buffer(1000)
#crear una nueva columna con el area de influencia de 10.000 metros
```

```python
municipios_extremadura_shp["buffer_centroide"] = municipios_extremadura_shp["centroide"].buffer(1000)
municipios_extremadura_shp["buffer_centroide"].plot(alpha = 0.5)
#generamos el buffer sombre el centroide
```

```python
axis = municipios_extremadura_shp["buffer"].plot(alpha = 0.5)
municipios_extremadura_shp["buffer_centroide"].plot(ax = axis, color = "red", alpha = 0.5)
municipios_extremadura_shp["perimetro"].plot(ax = axis, color = "white", linewidth = 0.5)
```

```python
#CREAR UN MAPA
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
# Cargar la capa temática
municipios_extremadura_shp = (r"C:\Users\molea\Documents\GIS DataBase\CartografiaExtremadura\municipios extremadura.shp")
map_data = gpd.read_file(municipios_extremadura_shp)
map_data.head(3)
```

```python
# Control del tamaño de la figura del mapa
fig, ax = plt.subplots(figsize=(5, 5))
 
# Control del título y los ejes
ax.set_title('Municipios Extremadura', 
             pad = 20, 
             fontdict={'fontsize':15, 'color': '#4873ab'})
ax.set_xlabel('Longitud')
ax.set_ylabel('Latitud')
 
# Mostrar el mapa finalizado
map_data.plot(column='NOMBRE03', cmap='plasma', ax=ax, zorder=5)
```
